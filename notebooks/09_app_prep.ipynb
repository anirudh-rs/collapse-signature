```python
import os
import json

required_files = [
    "../data/processed/civilisations_clean.csv",
    "../data/processed/external_stressors.csv",
    "../data/processed/collapse_clusters.csv",
    "../data/processed/network_centrality.csv",
    "../data/processed/shap_importance.csv",
    "../data/processed/shap_values.csv",
    "../data/processed/survival_probabilities.csv",
    "../data/processed/cluster_stressor_alignment.csv",
    "../data/processed/pillar_1_2_results.json",
    "../data/raw/civilisations_master.csv",
    "../data/raw/external_stressors.csv",
]

print("=== PRE-APP FILE CHECK ===\n")
all_good = True
for f in required_files:
    if os.path.exists(f):
        size = os.path.getsize(f) / 1024
        print(f"  ✓ {f.replace('../', '')} ({size:.1f} KB)")
    else:
        print(f"  ✗ MISSING: {f.replace('../', '')}")
        all_good = False

with open("../data/processed/pillar_1_2_results.json") as f:
    results = json.load(f)

print(f"\n=== KEY RESULTS CONFIRMED ===")
print(f"  Temporal clustering p-value:  {results['pillar1_p_value']}")
print(f"  Clusters identified:          {results['pillar1_n_clusters']}")
print(f"  Global Moran's I:             {results['pillar2_global_morans_i']}")
print(f"  Stressor alignment:           {results['pillar3_mean_alignment_score']:.0%}")
print(f"  Longest chain:                {results['pillar4_longest_chain_length']} civilisations")
print(f"  Model LOO accuracy:           {results['pillar5_loo_accuracy']:.1%}")
print(f"  Largest counterfactual jump:  {results['pillar5_largest_jump']:.1%}")
print(f"\n{'ALL FILES READY' if all_good else 'MISSING FILES — fix before proceeding'}")
```
